### Complimentary code for plotting

#### Installation of packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Permanent Error

```python
# x-axis values
x = [0, 0.15, 0.30]
# y-axis values
y = [19.416, 19.344, 19.344]

# plotting points as a scatter plot
plt.scatter(x, y, label = "e_ss", color = "red", marker = "o", s=30)

# x-axis label
plt.xlabel('Kd')
# frequency label
plt.ylabel('e_ss (deg)')
# plot title
plt.title('Depiction of permanent error for Kd with Kp = 1')
# showing legend
plt.legend()

linear_model=np.polyfit(x,y,1)
linear_model_fn=np.poly1d(linear_model)
x_s=np.arange(0, 1.2)
plt.plot(x_s,linear_model_fn(x_s),color="green")
 
# function to show the plot
plt.show()
```

#### Lifting time

```python
# x-axis values
x = [0, 0.15, 0.30]
# y-axis values
y = [0.18, 0.24, 0.54]

# plotting points as a scatter plot
plt.scatter(x, y, label = "T", color = "red", marker = "o", s=30)

# x-axis label
plt.xlabel('Kd')
# frequency label
plt.ylabel('T (sec)')
# plot title
plt.title('Depiction of lifting time for Kd with Kp = 1')
# showing legend
plt.legend()

linear_model=np.polyfit(x,y,1)
linear_model_fn=np.poly1d(linear_model)
x_s=np.arange(0, 1.2)
plt.plot(x_s,linear_model_fn(x_s),color="green")

# function to show the plot
plt.show()
```

#### Overshoot

```python
# x-axis values
x = [0, 0.15, 0.30]
# y-axis values
y = [0.1430, 0.0188, 0.0038]

# plotting points as a scatter plot
plt.scatter(x, y, label = "Overshoot", color = "red", marker = "o", s=30)

# x-axis label
plt.xlabel('Kd')
# frequency label
plt.ylabel('Overshoot')
# plot title
plt.title('Depiction of overshooting for Kd with Kp = 1')
# showing legend
plt.legend()

linear_model=np.polyfit(x,y,1)
linear_model_fn=np.poly1d(linear_model)
x_s=np.arange(0, 1.2)
plt.plot(x_s,linear_model_fn(x_s),color="green")
 
# function to show the plot
plt.show()
```

#### Critical damping

```python
# x-axis values
x = [1, 5, 20]
# x = [1, 5, 10, 20]

# y-axis values
y = [0.3, 0.6, 1.2]
# y = [0.3, 0.6, 0.6, 1.2]

# plotting points as a scatter plot
plt.scatter(x, y, label = "Real Values", color = "red", marker = "o", s=30)

# x-axis label
plt.xlabel('Kp')
# frequency label
plt.ylabel('Kd')
# plot title
plt.title('Linear approximation of Kp, Kd in critical damping')
# showing legend
plt.legend()

linear_model=np.polyfit(x,y,1)
linear_model_fn=np.poly1d(linear_model)
x_s=np.arange(0,20)
plt.plot(x_s,linear_model_fn(x_s),color="green")

# function to show the plot
plt.show()
```
